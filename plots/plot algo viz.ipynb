```python
import pandas as pd
import numpy as np
np.random.seed(42)
mem_age = 4

result = []

max_ = 0
for i in range(1, mem_age + 1):
    for j in range(10):
        err = abs(np.random.standard_gamma(6))
        max_ = err if err > max_ else max_
        result.append([err, i])
        

df_1 = pd.DataFrame(result).rename({0:"err", 1:"age"}, axis=1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_stuff(df):
    fig = plt.figure(figsize=(20,8))

    palette = sns.color_palette("Greys", 6, 1)
    colors = [palette[i] for i in df.age.values]
    
#     print(df.index.values, df.err.values, colors)

    plt.bar(df.index.values, df.err.values, color=colors)
    plt.tick_params(left=False, bottom=False)

    fig.get_axes()[0].xaxis.set_visible(False)
    fig.get_axes()[0].yaxis.set_visible(False)
    fig.get_axes()[0].spines['left'].set_visible(False)
    fig.get_axes()[0].spines['right'].set_visible(False)
    fig.get_axes()[0].spines['top'].set_visible(False)
    fig.get_axes()[0].spines['bottom'].set_visible(False)
plot_stuff(df_1)
```

```python
result = []
exp = lambda x: 0.5*np.exp(-0.5*x)

max_ = 0
for i in range(1, mem_age + 1):
    for j in range(10):
        # err = abs(np.random.standard_gamma(6))
        # max_ = err if err > max_ else max_
        x = exp((10*(i-1) + j)/6)
        result.append([x, np.abs(i-5)])
        

df_exp = pd.DataFrame(result).rename({0:"err", 1:"age"}, axis=1)
plot_stuff(df_exp)
```

```python
appendage = [[max_, 5] for i in range(10)]
df_2 = pd.concat([df_1, pd.DataFrame(appendage, columns=["err", "age"])], axis=0)
df_2 = df_2.reset_index(drop=True)

plot_stuff(df_2)
```

```python
df_3 = df_2.sort_values(by="err", kind="stable")
df_3 = df_3.reset_index(drop=True)
plot_stuff(df_3)
```
